```python
import seaborn as sns
```

```python
#basic plotting
tips = sns.load_dataset('tips')
```

```python
#scatter plot
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

sns.scatterplot(x = 'total_bill' , y = 'tip', data = tips)
plt.title("scatter plot of total bill by tips")
plt.show() 
```

```python
#line plot
sns.lineplot(x = 'size' , y = 'total_bill' , data = tips)
plt.title("line plot of total bill by size")
plt.show()
```

```python
#categorical plots
sns.barplot(x = 'day',y = 'total_bill',data = tips)
plt.title("bar plot of total bi;ll by day")
plt.show()
```

```python
# boxen plot:
sns.boxenplot(x = 'day',y = 'total_bill',data = tips)
plt.show()
```

```python
# box plot:
sns.boxplot(x = 'day',y = 'total_bill',data = tips)
plt.show()
```

```python
#voilen plot:
sns.violinplot(x = 'day',y = 'total_bill',data = tips)
plt.show()
```

```python
#histograms:
sns.histplot(tips['total_bill'],bins = 10,kde = True)
plt.show()
#KDE =  Kernel Density Estimate ,  It represents the data using a continuous probability density curve in one or more dimensions
```

```python
#KDE plot:
# sns.kdeplot(tips['total_bill'])
sns.kdeplot(tips['total_bill'],fill = True)
```

```python
#pair plot
sns.pairplot(tips)
''' here categorial variable is not taken in consideration '''
```

```python
''' KEEP IN MIND '''
corr = tips.corr()
corr
```

```python
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
# Heat map: gives plot of these correlations:
sns.heatmap(corr,annot=True,cmap = 'flare')
```

```python
#customization of plots:
import pandas as pd
sales_data = pd.read_csv('sales_data.csv')
sales_data.head()
```

```python
#plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x = 'Product Category',y='Total Revenue',data = sales_data,estimator=sum)
plt.title('Total sales by product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```
